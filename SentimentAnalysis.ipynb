```python
import numpy as np
import pandas as pd
import nltk
```

```python
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sentiment = SentimentIntensityAnalyzer()
```

```python
df = pd.read_csv("IMDB Dataset.csv")
df.head()
```

```python
df['scores'] = df['review'].apply(lambda body:sentiment.polarity_scores(str(body)))
df.head()
```

```python
df['compound'] = df['scores'].apply(lambda score_dict:score_dict['compound'])
df.head()
```

```python
df['pos']=df['scores'].apply(lambda pos_dict:pos_dict['pos'])
df.head()
```

```python
df['neg'] = df['scores'].apply(lambda neg_dict:neg_dict['neg'])
df.head()
```

```python
df['type'] = ''
df.loc[df.compound>0, 'type'] = 'POSITIVE'
df.loc[df.compound==0, 'type'] = 'NEUTRAL'
df.loc[df.compound<0, 'type'] = 'NEGATIVE'
df.head()
```

```python
df.tail()
```

```python
len = df.shape
(rows,cols)  = len 
pos = 0
neg = 0
neutral = 0
for i in range(0,rows):
    if df.loc[i][3]=='POSITIVE':
        pos=pos+1
    if df.loc[i][3]=='NEGATIVE':
        neg  = neg+1
    if df.loc[i][3]=='NEUTRAL':
        neutral = neutral +1
```

```python
print("Positive reviews: "+str(pos)+" Negative reviews: "+str(neg)+" Neutral reviews: "+str(neutral))
```
